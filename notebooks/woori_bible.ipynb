```python
import pandas as pd

# Open the file and read lines into a list
with open('woorimal.txt', 'r', encoding='cp949') as file:
    lines = file.read().splitlines()

# Convert list to DataFrame
df = pd.DataFrame(lines, columns=['verse'])
df.head()
```

```python
df[['verse', 'text']] = df['verse'].str.split('\t', expand=True)
df.head()
```

```python
pattern = r'(?P<book>\D+?)\s(?P<chapter>\d+):(?P<verse>\d+)'

# Use str.extract() to apply the pattern
df_extracted = df['verse'].str.extract(pattern)

# Show the DataFrame with the new columns
df_extracted.head()
```

```python
df['book_name_kr'] = df_extracted['book']
df['chapter'] = df_extracted['chapter']
df['verse'] = df_extracted['verse']
df = df[['book_name_kr', 'chapter', 'verse', 'text']]
df
```

```python
unique_books = list(df['book_name_kr'].unique())

print(unique_books)
```

```python
full_name = ['창세기', '출애굽기', '레위기', '민수기', '신명기', '여호수아', '사사기', '룻기', 
             '사무엘상', '사무엘하', '열왕기상', '열왕기하', '역대상', '역대하', '에스라', 
             '느헤미야', '에스더', '욥기', '시편', '잠언', '전도서', '아가서', '이사야', '예레미야',
             '예레미야애가', '에스겔', '다니엘', '호세아', '요엘', '아모스', '오바댜', '요나', 
             '미가', '나훔', '하바국', '스바냐', '학개', '스가랴', '말라기', '마태복음', '마가복음',
             '누가복음', '요한복음', '사도행전', '로마서', '고린도전서', '고린도후서', '갈라디아서', 
             '에베소서', '빌립보서', '골로새서', '데살로니가전서', '데살로니가후서', '디모데전서', 
             '디모데후서', '디도서', '빌레몬서', '히브리서', '야고보서', '베드로전서', '베드로후서', 
             '요한 1서', '요한 2서', '요한 3서', '유다서', '요한계시록']
```

```python
# Create a mapping dictionary from abbrev_names to full_names
name_mapping = dict(zip(unique_books, full_name))

# replace values
df['book_name_kr'] = df['book_name_kr'].replace(name_mapping)
```
